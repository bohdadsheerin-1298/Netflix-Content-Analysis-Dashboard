```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('netflix_titles.csv',encoding='latin-1')
print(df.head())

#removing duplicates
print(df.drop_duplicates(inplace=True))

#checking for null values
print(df.isnull().sum())

#handling missing values
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Nan')
df['release_year'] = df['release_year'].fillna('Nan')

#converting to datetime
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')

#year column added
df['year_added'] = df['date_added'].dt.year

#content category
df.rename(columns={'type': 'content_type'},inplace=True)
#df['content_type'] = df['type']

#extracted primary genre
df['primary_genre'] = df['listed_in'].str.split(',').str[0]

#duration in numeric
df['duration_minutes'] = df['duration'].str.extract(r'(\d+)').astype(float)


df.to_csv('cleaned_netflix_data.csv',index=False,encoding='utf-8')
```
